```python
import numpy as np
import matplotlib.pyplot as plt
distribution = (np.random.rand(1000)*100 + 5).astype(int) # распределение ближе к нормальному, тк все значения сосредоточены в центре
# чтобы распределение изменило форму, можно взять другие данные, сделать функцию не рандом например
center = np.ones(5) * 50
high_outliers = (np.random.rand(10)*200).astype(int)
low_outliers = (np.random.rand(10)*-100).astype(int)
series = np.concatenate((distribution, center, high_outliers, low_outliers))
pd.Series(series).plot.hist(bins=500) # параметры bins можно менять
plt.show()
```

```python
def frequency_use(a):
    l = len(a)
    b = list(range(l))
    for i in range(l):
      b[a[i]] += 1
ser = (np.random.rand(1000) * 100 + 5).astype(int)
frequency_use(ser)
```

```python
import numpy as np

def sum(a):
    s = 0
    for i in range(len(a)) :
      s += int(a[i])
    
    print("Сумма чисел равна {}\n". format(s))

series = (np.random.rand(1000) * 100 + 5).astype(int)  
sum(series)
```

```python
import numpy as np
def expectation(a):
    s = 0
    for i in range(len(a)):
      s += a[i] 
    print("Мат ожидание чисел равно {}\n". format(s / 1000))

series = (np.random.rand(1000) * 100 + 5).astype(int)  
expectation(series)
```

```python
import numpy as np

def disp(a):
    s = 0
    s1 = 0
    for i in range(len(a)):
      s += a[i]
      s1 += a[i] * a[i]
    print("Дисперсия чисел равна {}\n". format(s1 / 1000 - (s / 1000) * (s / 1000)))

series = (np.random.rand(1000) * 100 + 5).astype(int)  
disp(series)
```

```python
import numpy as np

def std(a):
    s = 0
    s1 = 0
    for i in range(len(a)):
      s += a[i]
      s1 += a[i] * a[i]
    print("Среднеквадратичное отклонение чисел на основании смещенной оценки дисперсии равно {}\n". 
          format((s1 / 1000 - (s / 1000) * (s / 1000)) ** 0.5))

series = (np.random.rand(1000) * 100 + 5).astype(int)  
std(series)
```

```python
import numpy as np

def std2(a):
    s = 0
    s1 = 0
    n = len(a)
    for i in range (n):
      s += a[i]
      s1 += a[i] * a[i]
    std = ((s1 / 1000 - (s / 1000) * (s / 1000)) ** 0.5)

    print("Среднеквадратичное отклонение чисел на основании несмещенной оценки дисперсии равно {}\n". 
          format( (n/(n-1)*std) ** 0.5))

series = (np.random.rand(1000) * 100 + 5).astype(int)  
std2(series)
```
